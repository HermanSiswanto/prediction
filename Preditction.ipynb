```python
import pandas as pd

df = pd.read_csv (r"C:\Users\herma\PythonEnvironment\Project4\heart_cleveland_upload.csv")
df.head(5) #menampilkan 5 data teratas
```

```python
#for filtering data
df2 = df[(df['age']>40)&(df['age']<60)]
df2
```

```python
x = df.iloc[:,0:13]
y = df.iloc[:,13]
x
```

33% of data will be in the test part and rest will be in train part.

```python
from sklearn.model_selection import train_test_split #fungsi library ini membagi  data training dan data testing
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=1)
```

```python
from sklearn.svm import SVC

clf = SVC()
clf = clf.fit(x_train, y_train) #fit untuk melakukan training
y_pred = clf.predict (x_test)
```

```python
from sklearn.metrics import accuracy_score

print (accuracy_score (y_pred, y_test))
```
